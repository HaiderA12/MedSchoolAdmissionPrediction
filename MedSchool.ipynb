```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib as plt
from sklearn import linear_model
from sklearn import preprocessing
```

```python
df_train = pd.read_csv('../MedSchool/Training.csv')
df_test = pd.read_csv('../MedSchool/Test.csv')
```

```python
#converting letter grades into numbers
WS_n = {"J":1,"K":2,"L":3,"M":4, "N":5, "O":6, "P":7, "Q":8, "R":9, "S":10, "T":11}
```

```python
df_train['WS'] = df_train['WS'].map(WS_n)
df_test['WS'] = df_test['WS'].map(WS_n)
```

```python
df_train = df_train[df_train['Answer key']!=3]
```

```python
df_train = df_train.drop(['MCAT Total'], axis = 1)
df_test = df_test.drop(['MCAT Total'], axis = 1)
```

```python
df_train[df_train['Answer key'] == 0]
```

```python
df_train['Undergrad Major'].describe()
```

```python
for i in df_train.index:
    if 'Bio' in df_train.loc[i,'Undergrad Major'] or 'Chem' in df_train.loc[i, 'Undergrad Major'] or 'Psy' in df_train.loc[i, 'Undergrad Major']:
        df_train.loc[i,'Bio_Chem_Psy'] = 1
    else:
        df_train.loc[i,'Bio_Chem_Psy'] = 0

for i in df_test.index:
    if 'Bio' in df_test.loc[i,'Undergrad Major'] or 'Chem' in df_test.loc[i, 'Undergrad Major'] or 'Psy' in df_test.loc[i, 'Undergrad Major']:
        df_test.loc[i,'Bio_Chem_Psy'] = 1
    else:
        df_test.loc[i,'Bio_Chem_Psy'] = 0
        
```

```python
df_train_x = df_train.reset_index().drop(['Student ID','Undergrad Major','Answer key','index' ], axis = 1)
```

```python
Y_train = df_train.reset_index().drop(['Student ID', 'Undergrad Major', 'CU BCPM GPA', 'CU AO GPA',
       'CU CUM GPA', 'VR', 'PS', 'WS', 'BS', 'Bio_Chem_Psy','index'],axis = 1)
```

```python
df_test_x = df_test.reset_index().drop(['Student ID', 'Undergrad Major','Probablitiy of acceptance','index' ], axis = 1)
```

```python
print(df_train_x.columns)
print(df_test_x.columns)
```

```python
#data = {'score': [234,24,14,27,-74,46,73,-18,59,160]}
#df = pd.DataFrame(data)
#df
min_max_scaler = preprocessing.MinMaxScaler()
#np_scaled_train = min_max_scaler.fit_transform(df_train_x)
#np_scaled_test =  min_max_scaler.fit_transform(df_test_x)

X_train = pd.DataFrame(min_max_scaler.fit_transform(df_train_x))
X_test = pd.DataFrame(min_max_scaler.fit_transform(df_test_x))
```

```python
#lINEAR REGRESSION
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,Y_train)
Y_pred = linreg.predict(X_test)
Y_pred2 = []
for i in Y_pred:
    temp = i[0]
    if temp < 0:
        temp = 0.001
    Y_pred2.append(temp)
```

```python
res = pd.DataFrame(df_test['Student ID'])
res['Probablitiy of acceptance'] = Y_pred2
res.to_csv('../MedSchool/result1.csv')
```

```python
test_x_norm2 = pd.DataFrame()
for i in df_test_x.columns:
    if i != 'Bio_Chem_Psy':
        test_x_norm2[i] = (df_test_x[i] - df_test_x[i].mean())/df_test_x[i].std()
test_x_norm2['Bio_Chem_Psy'] = df_test_x['Bio_Chem_Psy']

train_x_norm2 = pd.DataFrame()
for i in df_train_x.columns:
    if i != 'Bio_Chem_Psy':
        train_x_norm2[i] = (df_train_x[i] - df_train_x[i].mean())/df_train_x[i].std()
train_x_norm2['Bio_Chem_Psy'] = df_train_x['Bio_Chem_Psy']
```

```python
#from sklearn.linear_model import LinearRegression
linreg2 = LinearRegression()
linreg2.fit(train_x_norm2,Y_train)
Y2_pred = linreg.predict(test_x_norm2)
Y2_pred2 = []
for i in Y2_pred:
    temp2 = i[0]
    if temp2 < 0:
        temp2 = 0.001
    if temp2>1:
        temp2 = 0.999
    Y2_pred2.append(temp2)
```

```python
res2 = pd.DataFrame(df_test['Student ID'])
res2['Probablitiy of acceptance'] = Y2_pred2
res2.to_csv('../MedSchool/result2.csv')
```
